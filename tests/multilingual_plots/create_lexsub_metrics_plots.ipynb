```python
import seaborn as sns
import pandas as pd
```

```python


dataset = 'germeval'
res = []
for dataset in ['semeval', 'germeval', 'semdisfr']:
    df = pd.read_csv(dataset + '.csv')
    df['dpl'] = df['dpl'].fillna('+embs')
    df['multitoken'] = df['multitoken'].fillna('CONCAT').replace({'cwm': "WCM"})

    df['dataset'] = dataset
    res.append(df)
df = pd.concat(res)
```

```python
translations = {
    t1: t2 for t1,t2 in zip(list(df.dpl.unique())[7:13], list(df.dpl.unique())[1:7])
} | {
    t1: t2 for t1,t2 in zip(list(df.dpl.unique())[13:19], list(df.dpl.unique())[1:7])
}
print(translations)
df['dpl'] = df['dpl'].replace(translations)
```

```python
df
```

```python
df.loc[df['multitoken'] == 'WCM', 'multitoken'] = df.loc[df['multitoken'] == 'WCM', 'multitoken'] + ' multi'
df = df.rename(columns={'rec_10': 'Recall@10'})
```

```python
df
```

```python
import matplotlib.pyplot as plt
import matplotlib

font = {
        'size'   : 17}

matplotlib.rc('font', **font)
metric = 'Recall@10'
sns.color_palette("tab10")

g = sns.FacetGrid(
    df[df.dpl.isin(['+embs', '<T> (and also <M>)', '<T> (or even <M>)', '<T> and <M>', '<T> or <M>'])
    ].replace({'<T> (and also <M>)': '"and also"', '<T> (or even <M>)': '"or even"', '<T> and <M>': '"and"', '<T> or <M>': '"or"'})
    , col="dataset", sharex=False, legend_out=False)
g.map_dataframe(sns.barplot, y='dpl', x=metric, hue='multitoken', palette="tab10")
g.set_titles("{col_name}")
g.set_ylabels('')
g.add_legend()
g.legend.remove()
plt.savefig('lexsub metrics.png', dpi=150, bbox_inches='tight')
```
